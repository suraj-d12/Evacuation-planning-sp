```python
# Example DataFrame with 'Year' and 'Deaths per 10,000' columns
# df_india = pd.read_csv('path_to_your_csv_file.csv')

# Ensure that the 'Year' column in df_india matches the format in population_df
df_india = df_india.merge(population_df, on='Year', how='left')
```

```python
# Assuming 'Deaths per 10,000' is the name of the relevant column in your df_india
df_india['Deaths from Drought'] = df_india['Deaths per 100,000 people - Drought'] * df_india['Population'] / 10000

# Drop the 'Deaths per 10,000' column if needed
df_india = df_india.drop(columns=['Deaths per 100,000 people - Drought'])
```

```python
# Assuming 'Deaths per 10,000' is the name of the relevant column in your df_india
df_india['Deaths from Flood'] = df_india['Deaths per 100,000 people - Flood'] * df_india['Population'] / 10000

# Drop the 'Deaths per 10,000' column if needed
df_india = df_india.drop(columns=['Deaths per 100,000 people - Flood'])
```

```python
df_india['Deaths from Drought'] = df_india['Deaths from Drought'].astype(float)
df_india['Deaths-Drought'] = df_india['Deaths from Drought'].apply(lambda x: f'{x:.2f}')
```

```python
df_india['Deaths-TemperatureExtreme'] = df_india['Deaths per 100,000 people - Extreme temperature'] * df_india['Population'] / 10000
df_india = df_india.drop(columns=['Deaths per 100,000 people - Extreme temperature'])
df_india = df_india.drop(columns=['Deaths per 100,000 people - Volcanic activity'])
df_india = df_india.drop(columns=['Deaths per 100,000 people - Wildfire'])
df_india = df_india.drop(columns=['Deaths per 100,000 people - Dry mass movement'])
```

```python
df_india['Deaths-GLOF'] = df_india['Deaths per 100,000 people - Glacial lake outburst flood'] * df_india['Population'] / 10000
df_india = df_india.drop(columns=['Deaths per 100,000 people - Glacial lake outburst flood'])
df_india = df_india.drop(columns=['Deaths per 100,000 people - Fog'])
```

```python
df_india['Deaths-Earthquake'] = df_india['Deaths per 100,000 people - Earthquake'] * df_india['Population'] / 10000
df_india = df_india.drop(columns=['Deaths per 100,000 people - Earthquake'])
```

```python
df_india['Deaths-ExtremWeather'] = df_india['Deaths per 100,000 people - Extreme weather'] * df_india['Population'] / 10000
df_india = df_india.drop(columns=['Deaths per 100,000 people - Extreme weather'])
```

```python
df_india['Deaths-WetMassMovement'] = df_india['Deaths per 100,000 people - Wet mass movement'] * df_india['Population'] / 10000
df_india = df_india.drop(columns=['Deaths per 100,000 people - Wet mass movement'])
```

```python
# Prepare the DataFrame for plotting
import matplotlib.pyplot as plt
df_plot = df_india.groupby('Year').agg({
    'Deaths from Drought': 'sum',
    'Deaths from Flood': 'sum',
    'Deaths-Drought': 'sum',
    'Deaths-TemperatureExtreme': 'sum',
    'Deaths-GLOF': 'sum',
    'Deaths-Earthquake': 'sum',
    'Deaths-ExtremWeather': 'sum',
    'Deaths-WetMassMovement': 'sum'
}).reset_index()

# Set the index to Year for plotting
df_plot.set_index('Year', inplace=True)

# Create a bar plot
ax = df_plot.plot(kind='bar', stacked=True, figsize=(12, 6))

# Customizing the plot
plt.title('Total Deaths by Disaster Type Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.xticks(rotation=45)
plt.legend(title='Disaster Type', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```
